```python
import numpy as np
from sklearn.datasets import load_iris
iris = load_iris()
import pandas as pd
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
df.head()
```

```python
df.shape
```

```python
df['target'] = iris.target
```

```python
df.head()
```

```python
iris.target_names
```

```python
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
```

```python
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],edgecolors='green',marker='*')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],edgecolors='blue',marker='.')
plt.show()
```

```python
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],edgecolors='green',marker='*')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],edgecolors='blue',marker='.')
plt.show()
```

```python
X = df.drop(columns=['target','flower_name'],axis='columns')
X.head()
```

```python
y = df.target
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
